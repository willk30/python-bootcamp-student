# Part 2: Creating Your First Git Repository üìÅ

Let's create a simple repository to practice Git commands. We'll build a small Python project for statistical calculations - perfect for ISE/OR students!

````python
# Create a project directory
import os

# Clean up if directory exists from previous run
if os.path.exists('stats_project'):
    !rm -rf stats_project

# Create and enter project directory
!mkdir stats_project
os.chdir('stats_project')

# Initialize Git repository
!git init

print("\n‚úÖ Created new Git repository in 'stats_project' directory!")
print("üìÅ Current directory:", os.getcwd())
````

````python
# Create our first Python file
stats_code = '''"""Statistical Calculator for ISE/OR Students
A simple module for basic statistical calculations.
"""

import math

def mean(data):
    """Calculate the arithmetic mean of a list of numbers."""
    return sum(data) / len(data)

def variance(data):
    """Calculate the population variance of a list of numbers."""
    m = mean(data)
    return sum((x - m) ** 2 for x in data) / len(data)

def std_dev(data):
    """Calculate the population standard deviation."""
    return math.sqrt(variance(data))

# Test the functions
if __name__ == "__main__":
    test_data = [2, 4, 6, 8, 10]
    print(f"Test data: {test_data}")
    print(f"Mean: {mean(test_data):.2f}")
    print(f"Variance: {variance(test_data):.2f}")
    print(f"Std Dev: {std_dev(test_data):.2f}")
'''

# Write the file
with open('stats_calc.py', 'w') as f:
    f.write(stats_code)

print("‚úÖ Created stats_calc.py")
print("\nüìÑ File contents:")
!cat stats_calc.py
````

````python
# Check Git status - this is the most used Git command!
print("üîç Git Status (what's the current state?):")
print("=" * 50)
!git status

print("\nüí° Git sees our new file but it's not tracked yet!")
````

````python
# Stage the file (add to staging area)
!git add stats_calc.py

print("‚úÖ Added stats_calc.py to staging area")
print("\nüîç Git Status after staging:")
!git status

print("\nüí° Now the file is staged and ready to be committed!")
````

````python
# Make our first commit
!git commit -m "Add basic statistical functions"

print("\nüéâ Made your first commit!")
print("\nüìù View commit history:")
!git log --oneline
````

### üéØ Mini Exercise: Add More Functions

Let's practice the Git workflow by adding more statistical functions!

````python
# Add new functions to our file
additional_code = '''

def median(data):
    """Calculate the median of a list of numbers."""
    sorted_data = sorted(data)
    n = len(sorted_data)
    if n % 2 == 0:
        return (sorted_data[n//2 - 1] + sorted_data[n//2]) / 2
    else:
        return sorted_data[n//2]

def mode(data):
    """Calculate the mode (most frequent value)."""
    from collections import Counter
    counts = Counter(data)
    max_count = max(counts.values())
    modes = [val for val, count in counts.items() if count == max_count]
    return modes[0] if len(modes) == 1 else modes
'''

# Append to existing file
with open('stats_calc.py', 'a') as f:
    f.write(additional_code)

print("‚úÖ Added median and mode functions")

# Check what changed
print("\nüîç Git diff (what changed?):")
!git diff
````

````python
# Stage and commit the changes
!git add stats_calc.py
!git commit -m "Add median and mode functions"

print("‚úÖ Committed new changes!")
print("\nüìù Updated commit history:")
!git log --oneline
````

# Part 3: Working with Branches üå≥

Branches let you experiment without affecting your main code. Let's create a feature branch!

````python
# Create and switch to a new branch
!git checkout -b add-advanced-stats

print("\nüåø Current branches:")
!git branch

print("\nüí° The * indicates which branch you're currently on")
````

````python
# Add advanced statistical functions on the new branch
advanced_code = '''

def correlation(x, y):
    """Calculate Pearson correlation coefficient between two lists."""
    if len(x) != len(y):
        raise ValueError("Lists must have the same length")
    
    n = len(x)
    mean_x = mean(x)
    mean_y = mean(y)
    
    numerator = sum((x[i] - mean_x) * (y[i] - mean_y) for i in range(n))
    denominator = math.sqrt(sum((x[i] - mean_x)**2 for i in range(n)) * 
                           sum((y[i] - mean_y)**2 for i in range(n)))
    
    return numerator / denominator if denominator != 0 else 0

def confidence_interval(data, confidence=0.95):
    """Calculate confidence interval for the mean."""
    from scipy import stats
    n = len(data)
    m = mean(data)
    std_err = std_dev(data) / math.sqrt(n)
    
    # For simplicity, using z-score (assumes large sample)
    z_score = 1.96 if confidence == 0.95 else 2.58  # 95% or 99%
    
    margin = z_score * std_err
    return (m - margin, m + margin)
'''

with open('stats_calc.py', 'a') as f:
    f.write(advanced_code)

print("‚úÖ Added advanced statistical functions on feature branch")

# Commit on the feature branch
!git add stats_calc.py
!git commit -m "Add correlation and confidence interval functions"

print("\nüìù Commits on feature branch:")
!git log --oneline -n 3
````

````python
# Switch back to main branch
!git checkout main

print("‚úÖ Switched to main branch")
print("\nüîç Notice: The advanced functions aren't here yet!")
print("\nLast 20 lines of stats_calc.py on main:")
!tail -20 stats_calc.py
````

````python
# Merge the feature branch into main
!git merge add-advanced-stats

print("\n‚úÖ Merged feature branch into main!")
print("\nüìù Updated history showing the merge:")
!git log --oneline -n 4

print("\nüéâ The advanced functions are now in main branch!")
````
